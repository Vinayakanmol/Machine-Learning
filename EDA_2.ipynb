#Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
df=pd.DataFrame({
    'Marks':[50,80,82,95,800],
    'Time':[1,2,3,1.5,60]
})
df
```

```python
iso=IsolationForest(random_state=10,contamination=0.1)
iso.fit(df)
out=iso.predict(df)
```

```python
out
```

```python
df['Score']=out
df
```
